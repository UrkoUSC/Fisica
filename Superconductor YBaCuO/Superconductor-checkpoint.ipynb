```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from matplotlib import rcParams

rcParams['figure.dpi']=150
```

```python
#Cargamos los datos para la resistividad y temperatura obtenidos en el laboratorio:

data = np.loadtxt('datos.txt')
```

```python
temperatura = data[:,0]
resistividad = data[:,1]
```

```python
plt.scatter(temperatura,resistividad, s=0.01, c='red')



plt.xlabel('Temperatura [K]')
plt.ylabel('Resistividad  $[\Omega\cdot m$]')



plt.grid()
plt.savefig('resistividad_temperatura.pdf')
plt.show()
```

```python
#Vamos a estudiar la resistividad en la zona lineal en un intervalo de 30 [K], entre 270[K] y 240[K]

#Zona 270 [K]
t_lineal_1=(temperatura<=285.0237)
t_lineal_11=temperatura[t_lineal_1]
t_lineal_2=(t_lineal_11>=264.0312)

#Temperaturas para ajuste
temperatura_lineal=t_lineal_11[t_lineal_2]

r_lineal_1=(resistividad<=234.9556)
r_lineal_11=resistividad[r_lineal_1]
r_lineal_2=(r_lineal_11>=215.1943)

#Resistividad para ajuste
resistividad_lineal=r_lineal_11[r_lineal_2]


#graficamos la zona lineal escogida

plt.scatter(temperatura_lineal, resistividad_lineal, s=0.01,
           c='r')

plt.plot(temperatura_lineal,ajustelineal(temperatura_lineal,termino_indep,pendiente),
         'b',
        linewidth=0.5,
        label='Ajuste contribucin normal')

plt.xlabel('Temperatura [K]')
plt.ylabel('Resistividad  $[\Omega\cdot m$]')

plt.grid()

plt.legend()
plt.savefig('zona_lineal_resistividad_temperatura.pdf')

plt.show()
```

```python
#Realizamos un ajuste lineal de la forma R = a + b*T

def ajustelineal(T,a,b):
    return a + (b*T)

popt, pcov = curve_fit(ajustelineal, temperatura_lineal, resistividad_lineal)
perr = np.sqrt(np.diag(pcov))

termino_indep=popt[0]
s_termino_indep=perr[0]

pendiente=popt[1]
s_pendiente=perr[1]


print('a')
print(termino_indep)
print('sa')
print(s_termino_indep)
print('b')
print(pendiente)
print('sb')
print(s_pendiente)
```
